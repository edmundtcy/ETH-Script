```python
file_path = r"C:\Users\Love\Downloads\trade_history.csv"
df = pd.read_csv(file_path)
os.remove(file_path)
df['time'] = pd.to_datetime(df['time'])
df.tail(30)
```

### Historical Data

```python
class position:
    def __init__(self, time, dir, coin, sz, ntl):
        self.time = time
        self.dir = dir
        self.coin = coin
        self.sz = sz
        self.ntl = ntl
    def reduce_size(self, sz, ntl):
        self.sz = self.sz - sz
        self.ntl = self.ntl + ntl
    def increase_size(self, sz, ntl):
        self.sz = self.sz + sz
        self.ntl = self.ntl + ntl

open_position_value = []
closed_trade = []

for index, row in df.iterrows():
    if row['dir'] == 'Open Long' or row['dir'] == 'Open Short':
        found = False
        for trade in open_position_value:
            if trade.coin == row['coin'] and trade.dir == row['dir']:
                trade.increase_size(row['sz'], row['ntl'])
                found = True
                break
        if not found:
            open_position_value.append(position(row['time'], row['dir'], row['coin'], row['sz'], row['ntl']))

    elif row['dir'] == 'Close Long' or row['dir'] == 'Market Order Liquidation: Close Long':
        for trade in open_position_value:
            if trade.coin == row['coin'] and trade.dir == 'Open Long':
                trade.reduce_size(row['sz'], row['ntl'])
                if trade.sz < 0.1:
                    closed_trade.append([trade.time, trade.coin, row['time']-trade.time, trade.ntl])
                    open_position_value.remove(trade)
    
    elif row['dir'] == 'Close Short' or row['dir'] == 'Market Order Liquidation: Close Short':
        for trade in open_position_value:
            if trade.coin == row['coin'] and trade.dir == 'Open Short':
                trade.reduce_size(row['sz'], row['ntl'])
                if trade.sz < 0.1:
                    closed_trade.append([trade.time, trade.coin, row['time']-trade.time, trade.ntl])
                    open_position_value.remove(trade)

result = [{'time': row[0], 'coin': row[1], 'duration': row[2], 'ntl': row[3]} for row in closed_trade]
df_duration = pd.DataFrame(result)
df_duration
```

```python
start = [
    pd.Timestamp('2024-04-10 08:00:00'),
    pd.Timestamp('2024-04-03 08:00:00'),
    pd.Timestamp('2024-03-06 08:00:00'),
    pd.Timestamp('2024-02-28 08:00:00'),
    pd.Timestamp('2024-02-21 08:00:00'),
    pd.Timestamp('2024-02-14 08:00:00'),
    pd.Timestamp('2024-02-07 08:00:00')
]
end = [
    pd.Timestamp('2024-04-17 08:00:00'),
    pd.Timestamp('2024-04-10 08:00:00'),
    pd.Timestamp('2024-03-13 08:00:00'),
    pd.Timestamp('2024-03-06 08:00:00'),
    pd.Timestamp('2024-02-28 08:00:00'),
    pd.Timestamp('2024-02-21 08:00:00'),
    pd.Timestamp('2024-02-14 08:00:00')
]
total_exchange_volume = [
    1,
    9673806247,
    10823100000,
    15440000000,
    8650700000,
    7601300000,
    7573400000
]
points = [1,168,82,25,9,7,1]
total_volume = []
for i in range(len(start)):
    start_time = start[i]
    end_time = end[i]
    filtered_df = df[(df['time'] >= start_time) & (df['time'] < end_time)]
    volume = filtered_df['ntl'].sum()
    total_volume.append(volume)
    filtered_df = df_duration[(df_duration['time'] >= start_time) & (df_duration['time'] < end_time)]
    coin_volume = filtered_df.groupby('coin')['ntl'].sum().sort_values(ascending=False)
    average_duration = np.mean([duration.total_seconds() for duration in filtered_df['duration']])
    print(f"Time period: {start_time} - {end_time}")
    print("Coin\t\tVolume")
    for coin, volume in coin_volume.items():
        print(f"{coin}\t\t{volume}")
    print(f"Average duration: {average_duration} seconds")
    print(f"Points: {points[i]}")
    print(f"My volume: {total_volume[i]}")
    print(f"Exchange volume: {total_exchange_volume[i]}")
    print(f"My volume share: {total_volume[i]/total_exchange_volume[i] * 100}%")
    print(f"My theoretical points: {total_volume[i]/total_exchange_volume[i] * 1000000}")
    print(f"Volume per point: {total_volume[i]/points[i]}")
    print("-----------------------------------------------------------------------------")
```
